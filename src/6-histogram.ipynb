### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2)

# Line chart
x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart')

# Bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter)
for i in range(len(x_scatter)):
    axs[1, 0].text(x_scatter[i], y_scatter[i], f'({x_scatter[i]}, {y_scatter[i]})')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
plt.figure()
plt.hist(data, color='skyblue', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

plt.tight_layout()
plt.show()
```
